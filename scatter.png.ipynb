```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
df = pd.read_csv(url)

# Convert 'Year' to numeric
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Filter recession data
rec_data = df[df['Recession'] == 1]

# Scatter plot: Consumer Confidence vs Automobile Sales
plt.figure(figsize=(10, 6))
plt.scatter(rec_data['Consumer_Confidence'], rec_data['Automobile_Sales'], color='teal', alpha=0.7)

plt.xlabel('Consumer Confidence')
plt.ylabel('Automobile Sales')
plt.title('Impact of Consumer Confidence on Automobile Sales During Recession')
plt.grid(True)
plt.tight_layout()
plt.show()
```
