```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
df = pd.read_csv(url)

# Convert 'Year' to numeric
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Filter data for recession period
Rdata = df[df['Recession'] == 1]

# Calculate total sales volume by vehicle type during recession
VTexpenditure = Rdata.groupby('Vehicle_Type')['Automobile_Sales'].sum()

# Create a pie chart for vehicle type share during recession
plt.figure(figsize=(8, 6))

labels = VTexpenditure.index
sizes = VTexpenditure.values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

plt.title('Vehicle Type Share of Automobile Sales During Recession')
plt.axis('equal')  # Ensures pie chart is circular
plt.tight_layout()
plt.show()
```
