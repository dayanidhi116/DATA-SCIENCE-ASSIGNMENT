```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
df = pd.read_csv(url)

# Convert 'Year' to numeric
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Filter non-recession data
non_rec_data = df[df['Recession'] == 0]

# Bubble size based on Seasonality_Weight
size = non_rec_data['Seasonality_Weight']

# Create scatter plot (bubble chart)
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=non_rec_data,
    x='Month',
    y='Automobile_Sales',
    size='Seasonality_Weight',
    hue='Seasonality_Weight',
    sizes=(50, 500),
    legend=False,
    alpha=0.6
)

plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.title('Seasonality Impact on Automobile Sales (Non-Recession Period)')
plt.tight_layout()
plt.show()
```
